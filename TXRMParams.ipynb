```python
#### FILE DIALOG TO SELECT FILE  ####

%gui qt

import TXRMParams as tp
from pathlib import Path
from PySide6.QtWidgets import QFileDialog

def gui_fname(dir=None):
    """Select a file via a dialog and return the file name."""
    if dir is None: dir ='./'
    fname = QFileDialog.getOpenFileName(None, "Select data file...", 
                dir, filter="All files (*);; Tomo Files (*.txrm)")
    return fname[0]

def gui_savename(dir=None):
    """Select a file via a dialog and return the file name."""
    if dir is None: dir ='./'
    fname = QFileDialog.getSaveFileName(None, "Confirm output filename to save...", 
                dir, filter="All files (*);; CSV Files (*.csv)")
    return fname[0]

file = gui_fname()

# Using with_suffix()
outfile = str(Path(file).with_suffix(""))+"_Params.csv")



Nos = range(5550,5555)

table = tp.MakeTable(file,Nos)

if(table == None):
    print(f"Error creating table")
    
else:
    try:
        with open(outfile, "w") as text_file:
            text_file.write(table)
        print(f"Successfully wrote table to {outfile}")
    except IOError as e:
        print(f"Error writing to file: {e}")

```
